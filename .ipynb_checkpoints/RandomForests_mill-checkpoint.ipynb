```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("Output/Millennial.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.columns
```

```python
new2_df=df.drop(columns=["q7@", "state", "Unnamed: 0"])
new2_df
```

```python
new2_df["q32"].value_counts(dropna=False)
```

```python
indexNames = new2_df[new2_df['q32'] == 9 ].index
```

```python
new2_df.drop(indexNames , inplace=True)
```

```python
new2_df["q32"].value_counts(dropna=False)
```

```python
filterinfDataframe = new2_df[(new2_df['age'] > 25) & (new2_df['age'] < 39)]
filterinfDataframe
```

```python
filterinfDataframe.to_csv("Output/filtered_milli.csv",index=False)
```

```python
milli_filtered = pd.read_csv('Output/filtered_milli.csv')
print(milli_filtered.shape)
milli_filtered.head()
```

```python
null_counts = milli_filtered.isnull().sum()
print("Number of null values in each column:\n{}".format(null_counts))
```

```python
milli_filtered.dtypes.value_counts()
```

```python
milli_filtered.info()
```

```python
X=milli_filtered.loc[:,["educ","age","kidsa","employ","health","home","sex","income","q1","q2","q3","race1_2","q8a"]]
y=milli_filtered["q32"].values.reshape(-1, 1)
```

```python
print("Shape: ", X.shape, y.shape)
```

```python

from sklearn.model_selection import train_test_split

### BEGIN SOLUTION

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
rf = RandomForestClassifier(n_estimators=48)
rf = rf.fit(X_train, y_train)
```

```python
importances = rf.feature_importances_
print(importances)
```

```python
prediction = rf.predict(X_test)
prediction
```

```python
rf.score(X_test, y_test)
```
