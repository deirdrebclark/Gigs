```python
milli_filtered = pd.read_csv('Output/filtered_health.csv')
print(milli_filtered.shape)
milli_filtered.head()
```

```python
milli_filtered.dtypes.value_counts()
```

```python
milli_filtered.info()
```

```python
null_counts = milli_filtered.isnull().sum()
print("Number of null values in each column:\n{}".format(null_counts))
```

```python
milli_filtered=milli_filtered.dropna()
```

```python
X=milli_filtered.loc[:,["AGE_P","WRKMYR","ERNYR_P"]]
y=milli_filtered["PHSTAT"].values.reshape(-1, 1)
```

```python
print("Shape: ", X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

### BEGIN SOLUTION

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
# Create and score a decision tree classifier
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
importances = clf.feature_importances_
importances
```

```python
prediction = clf.predict(X_test)
```

```python
clf.score(X_test, y_test)
```
